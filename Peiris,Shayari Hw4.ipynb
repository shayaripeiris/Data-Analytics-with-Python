## Question 6

Make a line plot and show 'Sub-Category' with respect to the Sum of their 'Profits'.

Your X axis should have Sub-Categories and Y axis should have the sum of profits.

Make sure you label the plot (Title, XLabel, YLabel)

Hint: Use .plot function

```python
#Problem 6: making a line plot and show 'Sub-Category' with respect to the Sum of their 'Profits'.

Subcategory_Profit_Sum = df.groupby('Sub-Category', as_index = False)['Profit'].sum()
Subcategory_Profit_Sum.plot(kind='line', x='Sub-Category', y='Profit',
                figsize = (12,8),
                xlabel = 'Sub-Categories',
                ylabel = 'Sum of Profits (Dollars)',
                title = 'Sum of Profits Over Sub-Categories')
plt.xticks(Subcategory_Profit_Sum.index, Subcategory_Profit_Sum['Sub-Category'], rotation=60)
plt.show()
```

## Question 7

Make a barplot of Order Priority with their Average Shipping Cost

Your X axis should have different Order Priorities and Y axis should have Shipping Cost

Make sure you label your axis

Hint: Use sns.barplot()

```python
#Problem 7: creating new df 'A' with shipping cost grouped by order priority

A = df.groupby('Order Priority', as_index = False)['Shipping Cost'].mean()
A
```

```python
#Problem 7 cont: making a barplot of Order Priority with their Average Shipping Cost
#Code:
plt.figure(figsize = (12,8))
sns.barplot("Order Priority", "Shipping Cost", data=A)
plt.xlabel('Order Priority')
plt.ylabel('Average Shipping Cost (Dollars)')
plt.title('Average Shipping Cost By Order Priority')
plt.show()
```

## Question 8

Make a scatter plot showing Quantity (X Axis) and Profit (Y Axis)

Use alpha = 0.7 and color = blue

Hint: Use df.plot.scatter()

```python
#Problem 8: making a scatter plot showing Quantity (X Axis) and Profit (Y Axis)

#Code:
df.plot.scatter(x = 'Quantity', y = 'Profit', c = 'blue', alpha=.7, figsize = (12,12));
plt.xlabel('Quantity')
plt.ylabel('Profits')
plt.title('Quantity vs. Profits Superstore Data')
plt.grid()
plt.show()
```

## Question 9

Make a boxplot for showing Ship Mode and Shipping Cost

The X axis should be Ship Mode and Y axis should be the Shipping Cost

Hint: Use sns.boxplot()

```python
#Problem 9: making a boxplot for showing Ship Mode and Shipping Cost
#Code:
plt.figure(figsize = (12,8))
sns.boxplot(x="Ship Mode", y="Shipping Cost", data=df, palette="Set2")   
plt.xlabel('Ship Mode')
plt.ylabel('Shipping Cost')
plt.title('Ship Mode vs. Shipping Cost')
plt.grid()
plt.show()
```

## Question 10

#Make a heatmap for checking the correlation between numerical variables

```python
#Problem 10: making a heatmap for checking the correlation between numerical variables
#Code:
plt.figure(figsize = (12,8))
corrplot = sns.heatmap(df[['Sales', 'Quantity', 'Discount', 'Profit','Shipping Cost', 'Cost']].corr())
corrplot.set_xticklabels(corrplot.get_xticklabels(), rotation = 45, horizontalalignment = 'right')
plt.title('Correlations Between Numerical Variables')
plt.show()
```
